```python
import codecs
import os
import nltk
from nltk.tokenize import MWETokenizer
tokenizer = nltk.data.load("tokenizers/punkt/english.pickle")
file=open("data.txt", mode='r',encoding="utf8");
#sentencelist=file.readlines();
content=file.read()

content=content.replace("i.e.", "aaaaaa");
content=content.replace("Fig.", "bbbbbb");
content=content.replace(" al.", "cccccc");
content=content.replace("Eqs.", "dddddd");
content=content.replace("e.g.", "eeeeee");

#print(type(content))
print(content)
```

```python
content='\n'.join(tokenizer.tokenize(content));
print(content)
```
